```python
import numpy as np 
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns
```

```python
data = load_iris()
```

```python
df = pd.DataFrame(data= data.data , columns= data.feature_names)
target = pd.DataFrame(data=data.target,columns = ["target"])
df
```

```python
df=pd.concat([df,target],axis=1)
df
```

```python
l =[]
for i in range(len(df)):
    if df.iloc[i,-1] == 2:
        l.append(i)
for i in range(len(l)-1,-1,-1):
    df=df.drop(df.index[l[i]])
    target=target.drop(target.index[l[i]])
```

```python
df.drop(["target"],axis=1,inplace=True)
df
```

```python
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split as tts
```

```python
x = np.array(df)
y = np.array(target)
x.shape,y.reshape(y.shape[0],)
```

```python
x_train,x_test,y_train,y_test = tts(x,y,train_size=0.7,shuffle=True,random_state=100)
```

```python
model = Perceptron()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred=y_pred.reshape(y_pred.shape[0],1)
```

```python
c=0
for i in range(len(y_test)):
    if y_test[i]==y_pred[i]:
        c+=1
print("Accuracy : ", c/len(y_test)*100)
```
